#### Ejemplos ciclo while

```python
# 1. Contador simple (Imprimir números del 1 al 5)
# Este es el uso más básico, donde una variable de control se incrementa en cada iteración.
i = 1              # Inicializa la variable de control en 1 [3].
while i <= 5:      # Evalúa si i es menor o igual a 5 [3].
    print(i)       # Muestra el valor actual de i [3].
    i += 1         # Incrementa i en 1 para evitar un bucle infinito [3].
```

```python
# 2. Cálculo del factorial de un número
# Este ejemplo muestra cómo realizar cálculos acumulativos dentro de un ciclo.
numero = 5         # Número del cual queremos el factorial [4].
factorial = 1      # Variable para acumular el producto [4].

while numero > 0:       # El ciclo sigue mientras numero sea mayor a 0 [4].
    factorial *= numero # Multiplica el acumulado por el número actual [4].
    numero -= 1         # Decrementa el número en 1 [4].

print(f"El factorial es: {factorial}") # Muestra el resultado final (120) [4].
```

```python
# 3. Suma de números con valor centinela (Bucle hasta número negativo)
# Se utiliza un ciclo infinito (while True) que se rompe con la instrucción break cuando se introduce un valor específico que indica el fin de la entrada
# .
suma = 0           # Inicializa el acumulador de la suma [4].

while True:        # Crea un bucle que se ejecuta indefinidamente hasta un break [4].
    numero = int(input("Ingresa un número (negativo para salir): ")) # Pide un dato [4].
    if numero < 0: # Condición de salida: si el número es negativo [4].
        break      # Sale del ciclo inmediatamente [4].
    suma += numero # Suma el número ingresado al total [4].

print(f"La suma total es: {suma}") # Muestra el total acumulado [4].
```

```python
# 4. Búsqueda de un múltiplo (Encontrar el primer múltiplo de 7 mayor que 50)
# Útil cuando no sabes cuántas veces se repetirá el ciclo, sino que buscas que se cumpla un estado específico.
numero = 51             # Comienza la búsqueda desde 51 [4].

while numero % 7 != 0:  # Se repite mientras el residuo de dividir por 7 no sea 0 [4].
    numero += 1         # Prueba con el siguiente número [4].

print(f"El primer múltiplo de 7 mayor que 50 es: {numero}") # Resultado: 56 [5].
```

```python
# 5. Validación de entrada (Adivinar un número)
# Este ejemplo es común en programas interactivos para forzar al usuario a ingresar el dato correcto
numero_secreto = 7      # Define el número que el usuario debe adivinar [6].
intento = 0             # Inicializa la variable para recibir la entrada.

while intento != numero_secreto: # Repite mientras el intento sea incorrecto [6].
    intento = int(input("Adivina el número secreto: ")) # Solicita entrada [6].
    if intento != numero_secreto:
        print("Incorrecto, intenta de nuevo.")

print("¡Felicidades! Adivinaste el número.") # Mensaje final tras salir del ciclo [6]
```
